```python
import pandas as pd
import matplotlib.pyplot as plt

# 
file_path = 'data/user_data/users_combined_info_500.csv'
data = pd.read_csv(file_path)

# 
country_distribution = data['country'].value_counts()

# 
plt.figure(figsize=(12, 8))
country_distribution.plot(kind='bar')
plt.title('Country and Region Distribution of Users')
plt.xlabel('Country')
plt.ylabel('Number of Users')
plt.xticks(rotation=45)
plt.tight_layout()

# 
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

# 
file_path = 'data/user_data/users_combined_info_500.csv'
data = pd.read_csv(file_path)

# 
data['event_time'] = pd.to_datetime(data['event_time'])

# 
data['timezone'] = data['event_time'].dt.tz

# 
timezone_distribution = data['timezone'].value_counts()

# 
plt.figure(figsize=(12, 8))
timezone_distribution.plot(kind='bar')
plt.title('User Timezone Distribution')
plt.xlabel('Timezone')
plt.ylabel('Number of Users')
plt.xticks(rotation=45)
plt.tight_layout()

# 
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 
file_path = 'data/user_data/users_combined_info_500.csv'
data = pd.read_csv(file_path)

# 
submission_counts = data['user_id'].value_counts()

# 
plt.figure(figsize=(12, 8))
submission_counts.plot(kind='bar')
plt.title('Submission Frequency by User')
plt.xlabel('User ID')
plt.ylabel('Number of Submissions')
plt.xticks(rotation=45)
plt.tight_layout()

# 
plt.show()

# 
# 
high_activity_threshold = submission_counts.mean()
high_active_users = submission_counts[submission_counts > high_activity_threshold]
low_active_users = submission_counts[submission_counts <= high_activity_threshold]

print("High Active Users:")
print(high_active_users)

print("\nLow Active Users:")
print(low_active_users)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 
file_path = 'data/user_data/users_combined_info_500.csv'
data = pd.read_csv(file_path)

# 
data['event_time'] = pd.to_datetime(data['event_time'])

# 
data['hour'] = data['event_time'].dt.hour

# 
hourly_activity = data['hour'].value_counts().sort_index()

# 
plt.figure(figsize=(12, 8))
hourly_activity.plot(kind='line', marker='o')
plt.title('User Activity by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Events')
plt.xticks(range(24))
plt.tight_layout()

# 
plt.show()
```
